### Monte Carlo: Prevendo o Lucro Bruto

```python
# Importando as bibliotecas necessárias:
import numpy as np
import matplotlib.pyplot as plt
```

Simulando as Receitas:

```python
# Informando os valores:
receita_esperada = 170
desv_pad_receita = 20
simulacoes = 1000
```

```python
# Simulação de receitas futuras:
receita = np.random.normal(receita_esperada,desv_pad_receita,simulacoes)
# np.random.normal() - função do numpy que gera aleatoriamente distribuições normais, baseado no valor medio informado e em seu desvio-padrao, repetindo varias vezes
receita
```

```python
# Gerando o gráfico da distribuição normal da receita:
from turtle import color


plt.figure(figsize=(15,6))
plt.plot(receita)
plt.axhline(y=receita_esperada+desv_pad_receita, color='red')
plt.axhline(y=receita_esperada, color='red')
plt.axhline(y=receita_esperada-desv_pad_receita, color='red')
plt.show()
```

Simulando o CMV:

```python
# Gerando os CMV:
cmv = - (receita * np.random.normal(0.6,0.1))
# '-' - como o CMV é um gasto, se coloca - na frente
# receita - valores simulados da receita dos quais serão calculados os CMV
# np.random.normal() - função do numpy que gerará as simulações do CMv
# 0.6 - media historica do CMV
# 0.1 - desvio-padrao historico do CMV
# Serão atribuídos CMVs aleatórios a cada uma das receitas simuladas
cmv
```

```python
# Gerando o gráfico da distribuição normal do CMV:
from turtle import color

media = cmv.mean() # media dos CMV gerados
desvio = cmv.std() # desvio-padrão dos CMV gerados

plt.figure(figsize=(15,6))
plt.plot(cmv)

plt.axhline(y=media+desvio, color='red')
plt.axhline(y=media, color='red')
plt.axhline(y=media-desvio, color='red')
plt.show()
```

Simulando o Lucro Bruto:

```python
# Gerando o Lucro Bruto:
lucro_bruto = receita + cmv
# receita - valores para a receita simulados
# cmv - valores simulados para o CMv
lucro_bruto
```

```python
# Cálculo do Maior Potencial de Lucro Bruto:
maior_lucro = max(lucro_bruto)
```

```python
# Cálculo do Menor Potencial de Lucro Bruto
menor_lucro = min(lucro_bruto)
```

```python
# Cálculo da Média do Lucro Bruto:
media_lucro = lucro_bruto.mean()
```

```python
# Cálculo do Desvio-padrão do Lucro Bruto:
desvio_lucro = lucro_bruto.std()
```

```python
# Plotando o gráfico dos Lucros Brutos Simulados:
plt.figure(figsize=(15,6))
plt.plot(lucro_bruto)
plt.axhline(y=maior_lucro, color='green')
plt.axhline(y=media_lucro+desvio_lucro, color='orange')
plt.axhline(y=media_lucro, color='red')
plt.axhline(y=media_lucro-desvio_lucro, color='orange')
plt.axhline(y=menor_lucro, color='green')
plt.show()
```

Plotando Histogramas com os Resultados:
