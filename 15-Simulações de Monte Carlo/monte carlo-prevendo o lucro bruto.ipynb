### Monte Carlo: Prevendo o Lucro Bruto

```python
# Importando as bibliotecas necess√°rias:
import numpy as np
import matplotlib.pyplot as plt
```

Simulando as Receitas:

```python
# Informando os valores:
receita_esperada = 170
desv_pad_receita = 20
simulacoes = 1000
```

```python
# Simula√ß√£o de receitas futuras:
receita = np.random.normal(receita_esperada,desv_pad_receita,simulacoes)
# np.random.normal() - fun√ß√£o do numpy que gera aleatoriamente distribui√ß√µes normais, baseado no valor medio informado e em seu desvio-padrao, repetindo varias vezes
receita
```

```python
# Gerando o gr√°fico da distribui√ß√£o normal da receita:
from turtle import color


plt.figure(figsize=(15,6))
plt.plot(receita)
plt.axhline(y=receita_esperada+desv_pad_receita, color='red')
plt.axhline(y=receita_esperada, color='red')
plt.axhline(y=receita_esperada-desv_pad_receita, color='red')
plt.show()
```

Simulando o CMV:

```python
# Gerando os CMV:
cmv = - (receita * np.random.normal(0.6,0.1))
# '-' - como o CMV √© um gasto, se coloca - na frente
# receita - valores simulados da receita dos quais ser√£o calculados os CMV
# np.random.normal() - fun√ß√£o do numpy que gerar√° as simula√ß√µes do CMv
# 0.6 - media historica do CMV
# 0.1 - desvio-padrao historico do CMV
# Ser√£o atribu√≠dos CMVs aleat√≥rios a cada uma das receitas simuladas
cmv
```

```python
# Gerando o gr√°fico da distribui√ß√£o normal do CMV:
from turtle import color

media = cmv.mean() # media dos CMV gerados
desvio = cmv.std() # desvio-padr√£o dos CMV gerados

plt.figure(figsize=(15,6))
plt.plot(cmv)

plt.axhline(y=media+desvio, color='red')
plt.axhline(y=media, color='red')
plt.axhline(y=media-desvio, color='red')
plt.show()
```

Simulando o Lucro Bruto:

```python
# Gerando o Lucro Bruto:
lucro_bruto = receita + cmv
# receita - valores para a receita simulados
# cmv - valores simulados para o CMv
lucro_bruto
```

```python
# C√°lculo do Maior Potencial de Lucro Bruto:
maior_lucro = max(lucro_bruto)
```

```python
# C√°lculo do Menor Potencial de Lucro Bruto
menor_lucro = min(lucro_bruto)
```

```python
# C√°lculo da M√©dia do Lucro Bruto:
media_lucro = lucro_bruto.mean()
```

```python
# C√°lculo do Desvio-padr√£o do Lucro Bruto:
desvio_lucro = lucro_bruto.std()
```

```python
# Plotando o gr√°fico dos Lucros Brutos Simulados:
plt.figure(figsize=(15,6))
plt.plot(lucro_bruto)
plt.axhline(y=maior_lucro, color='green')
plt.axhline(y=media_lucro+desvio_lucro, color='orange')
plt.axhline(y=media_lucro, color='red')
plt.axhline(y=media_lucro-desvio_lucro, color='orange')
plt.axhline(y=menor_lucro, color='green')
plt.show()
```
